```python
# DEFINE AN IMAGE SIZE:
image_size=224

# CREATE AN IMAGE FOR PREPROCESSING:
def process_image(Image_Path,Image_Size=224):
    
    # READ AN IMAGE FILE
    Image=tf.io.read_file(Image_Path)
    # TURN THE JPEG IMAGE INTO NUMERICAL TENSOR WITH 3 COLOR CHANNEL
    Image=tf.image.decode_jpeg(Image,channels=3)
    # CONVERT THE COLOR CHANNEL VALUES FROM 0-255 TO 0-1
    Image=tf.image.convert_image_dtype(Image,tf.float32)
    # RESIZE THE IMAGE TO THE DESIRED VALUE (244,244)
    Image=tf.image.resize(Image,size=[Image_Size,Image_Size])
                          
    return Image
```

```python
# TAKES AN IMAGE FILE PATH NAME AND THE ASSOCIATED LABEL,PROCESS AN IMAGE AND RETURNS A TYPE OF TUPLE OF(IMAGE,LABEL)
def get_image_label(Image_Path,Boolean_Breeds):
    Image=process_image(Image_Path)
    return Image,Boolean_Breeds
```
